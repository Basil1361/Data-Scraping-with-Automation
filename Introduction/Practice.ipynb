Task: Wikipedia Search Validation
Goal: Write a Selenium script that searches Wikipedia and validates the results.

Requirements:

Open Wikipedia (https://www.wikipedia.org)
Verify "Wikipedia" appears in the page title
Find the search input field and search for "Artificial Intelligence"
Press Enter to perform the search
Add a wait so the page loads (use time.sleep(3))
Add assertions to verify:
The article page loaded successfully
The title contains "Artificial Intelligence"
The word "intelligence" appears somewhere in the page source
The page does NOT show "Wikipedia does not have an article with this exact name"
Use try/except to catch errors and print a message if something goes wrong
Hints:

The search box on Wikipedia's homepage has name="search"
After searching, check both driver.title and driver.page_source
Use .lower() for case-insensitive checks when needed
Don't forget time.sleep() after pressing Enter!
Bonus Challenge (if you finish):

Add a second search for gibberish like "xyzabc123notreal"
Write a try/except that expects the "not found" message this time

```python
# Solution
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.wikipedia.org")
assert "Wikipedia" in driver.title

element = driver.find_element(By.NAME, "search")
element.clear()
element.send_keys("xyzabc123notreal")
element.send_keys(Keys.RETURN)

time.sleep(3)

try:
    # Can add for error case also:
    assert "Error" not in driver.page_source
    case_lower = "Artificial Intelligence"
    assert case_lower.lower() in driver.page_source
    print("Artificial Intelligence is in Page")
    
except AssertionError:
    print("No_Results")
    print("Taking a screenshot")
    driver.save_screenshot("image.png")
```

Task: Google Search & Navigate Back
Goal: Practice basic navigation and page interaction.

Requirements:

Open Google (https://www.google.com)
Verify "Google" appears in the page title
3.1 - Interact: Find the search box and click it
3.2 - Fill form: Type "Python programming" into the search box
Press Enter to search
Wait for results to load (time.sleep(2))
Add assertions to verify:
The page title now shows "Python programming"
The word "python" appears in the page source
The page is NOT showing an error
3.6 - Navigate: Go back to the previous page using driver.back()
Verify you're back at the Google homepage
Use try/except to catch errors and print messages
Hints:

Google search box has name="q"
Use element.send_keys() to type text
Use Keys.RETURN to press Enter
Use driver.back() to go back
Use time.sleep() to wait between actions
Bonus: After going back, take a screenshot!

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Firefox()
driver.get("https://duckduckgo.com/")
element = driver.find_element(By.NAME, "q")
element.send_keys("Iceman lyrics")
time.sleep(3)
for _ in range(2):
    element.send_keys(Keys.ARROW_DOWN)
# element.send_keys(Keys.ARROW_DOWN, Keys.ARROW_DOWN) also works
time.sleep(3)
element.send_keys(Keys.RETURN)
# driver.find_element(By.ID, "search_bar")

# Verify if it's correct:
try:
    str = "Iceman"
    assert str.lower() in driver.page_source
    print("Correct searches")
    
except AssertionError:
    print("No Results, sent screenshot")
    driver.save_screenshot("image.png")    
```
