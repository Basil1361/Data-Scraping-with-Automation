```python
from selenium import webdriver
# selenium.webdriver module provides all the WebDriver implementations. 
from selenium.webdriver.common.keys import Keys
# The Keys class provide keys in the keyboard like RETURN, F1, ALT etc.
from selenium.webdriver.common.by import By
# The By class is used to locate elements within a document.
import time

driver = webdriver.Firefox()
# Creates a Firefox browser instance you'll control
driver.get("http://www.python.org")
# Opens python.org
assert "Python" in driver.title
# check python is in the page title
elem = driver.find_element(By.NAME, "q")
# Search bar
elem.clear()
# Clear search bar
elem.send_keys("shgsrgBSgSBSgrsg")
# enter pycon
elem.send_keys(Keys.RETURN)
# Enter


time.sleep(3)
try:
    # Try to do this:
    assert "No results found." not in driver.page_source
    # this means if "No results found." is *not* in the page
    print("✓ Search worked! Found results!")
    
except AssertionError:
    # if "No results found." is in the page
    print("❌ SOMETHING WENT WRONG! No results found on the page!")
    print("I'm taking a screenshot for you...")
    driver.save_screenshot("error.png")
```
